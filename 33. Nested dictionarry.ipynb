```python
'''33. Write a code to access a value in a nested dictionary. The function should take the dictionary
and a list of keys as input, and return the corresponding value. If any of the keys do not exist in
dictionary, the function should return none.'''
nested_dict = {
    'a': {
        'b': {
            'c': 'd'
        }
    },
    'e': {
        'f': 'g'
    }
}

keys = ['a', 'b', 'c']
current_value = nested_dict

for key in keys:
    if isinstance(current_value, dict) and key in current_value:
        current_value = current_value[key]
    else:
        current_value = None
        break

print(f"The value for keys {keys} is: {current_value}")

keys = ['e', 'f']
current_value = nested_dict

for key in keys:
    if isinstance(current_value, dict) and key in current_value:
        current_value = current_value[key]
    else:
        current_value = None
        break

print(f"The value for keys {keys} is: {current_value}")

keys = ['a', 'x', 'y']
current_value = nested_dict

for key in keys:
    if isinstance(current_value, dict) and key in current_value:
        current_value = current_value[key]
    else:
        current_value = None
        break

print(f"The value for keys {keys} is: {current_value}")
```
