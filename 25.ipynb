```python
"""25. Create a code that takes a tuple and two integers as input. The function should return
 a new tuple containing elements from the original tuple within the specific range of indices"""
tpl = (input("Enter a tuple: "))
tpl = tuple(tpl.split(','))
str1 = int(input("Enter interger values:"))
str2 = int(input("Enter interger values:"))
if str1 < 0 or str2 >= len(tpl) or str1>str2:
    print("Indices are out of range.")
else:
    new_tuple = tpl[str1:str2 + 1]
    print("Original Tuple:", tpl)
    print("New Tuple:", new_tuple)
```
