```python
#27. Develop a code that takes a tuple of integers as input. The function should return that maximum and minimum values from the tuple using tuple unpacking.
str_lst = input("Enter integer values seprated by commas: ")
int_lst = str_lst.split(",")
tpl = tuple(map(int,int_lst))
min_tpl = min(tpl)
max_tpl = max(tpl)
print("Minimum value is :", min_tpl)
print("Maximum value is :", max_tpl)



```
