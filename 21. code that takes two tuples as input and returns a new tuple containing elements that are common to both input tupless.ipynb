```python
#21. Write a code that takes two tuples as input and returns a new tuple containing elements that are common to both input tupless.
tuple1 = (1,2,3,4,5)
tuple2 = (1,3,5,6)
common_elements = []
for i in tuple1:
  if i in tuple2:
    common_elements.append(i)
same = tuple(common_elements)
print(same)
```
