```python
'''31. Write a code that takes a list of words as input and returns a dictionery
where the keys are unique words and the values are the frequinces of those words in the input list.'''
lst_tem = input("Enter the list:")
lst = set(lst_tem.split(","))
words_frqcs = {}
for word in lst:
  words = word.split()
  for i in words:
    if i in words_frqcs:
      words_frqcs[i] += 1
    else:
      words_frqcs[i] = 1
print(words_frqcs)
```
